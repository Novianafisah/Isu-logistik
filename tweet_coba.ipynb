```python
import twint
import nest_asyncio
nest_asyncio.apply()
```

```python
import pandas as pd
```

```python
column_names = ['date','time','username','tweet','Geo','link','likes','retweets','favorites','replies','mentions']
df = pd.DataFrame(columns = column_names)
keyword_list = []
#keyword1 = ["indonesia","ekspor","erick thohir","lockdown","pandemi","perdagangan internasional","antrian kapal"]
#keyword2 = ["kapasitas vessel","kendala logistik","pelabuhan","supply chain","krisis logistik global","kelangkaan kapal"]
keyword1 = ["indonesia","ekspor","erick thohir","lockdown","pandemi","perdagangan internasional","antrian kapal"]
keyword2 = ["kapasitas vessel","kendala logistik","pelabuhan","supply chain","krisis logistik global","kelangkaan kapal"]


print("List comprehension:")
for x, y in [(x,y) for x in keyword1 for y in keyword2]:
    keyword_list.append(x)
    keyword_list.append(y)
print(keyword_list)

for i in range(0, len(keyword_list), 2):
  keyword = keyword_list[i:i+2]
  
  c = twint.Config()

  c.Search = keyword
  c.Since = "2020-01-01"
  c.Until = "2021-11-30"
  c.Limit = 10000
  c.Pandas = True
  c.Custom = ['date','time','username','tweet','Geo','link','likes','retweets','favorites','replies','mentions']

  twint.run.Search(c)
  tweets_df = twint.storage.panda.Tweets_df
  df = pd.concat([df, tweets_df])
```

```python
column_names = ['date','time','username','tweet','Geo','link','likes','retweets','favorites','replies','mentions']
df_eng = pd.DataFrame(columns = column_names)
keyword_list = []
keyword1 = ["export","lockdown","pandemic","international trade","ship queue"]
keyword2 = ["vessel capacity","logistical problems","port","supply chain","global logistic crisis","ship scarcity"]

print("List comprehension:")
for x, y in [(x,y) for x in keyword1 for y in keyword2]:
    keyword_list.append(x)
    keyword_list.append(y)
print(keyword_list)

for i in range(0, len(keyword_list), 2):
  keyword = keyword_list[i:i+2]
  
  c = twint.Config()

  c.Search = keyword
  c.Since = "2020-01-01"
  c.Until = "2021-11-30"
  c.Limit = 10000
  c.Pandas = True
  c.Custom = ['date','time','username','tweet','Geo','link','likes','retweets','favorites','replies','mentions']

  twint.run.Search(c)
  tweets_df_eng = twint.storage.panda.Tweets_df
  df_eng = pd.concat([df_eng, tweets_df_eng])
```
